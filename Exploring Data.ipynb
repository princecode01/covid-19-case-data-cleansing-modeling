### Data Exploration

```python
from pathlib import Path
import pandas as pd

data_path = Path("data/raw")
csv_files = list(data_path.glob("*.csv"))
```

#### Dataset Metadata Summary

```python
summary = []

for file in csv_files:

    try:
        df = pd.read_csv(file)

        summary.append({
            "file": file.name,
            "rows": df.shape[0],
            "columns": df.shape[1],
            "column_names": ", ".join(df.columns),
            "missing_values": df.isnull().sum().sum(),
            "duplicates": df.duplicated().sum()
        })

    except Exception as e:
        summary.append({
            "file": file.name,
            "error": str(e)
        })

summary_df = pd.DataFrame(summary)

print(summary_df.head())

summary_df.to_csv("dataset_summary.csv", index=False)
```

#### Count Unique Schemas

```python
schemas = {}

for file in csv_files:

    try:
        df = pd.read_csv(file)

        cols = tuple(sorted(df.columns))

        if cols not in schemas:
            # Add the new schema to the dictionary
            schemas[cols] = []

        schemas[cols].append(file.name)

    except Exception as e:
        print(file.name, e)

print(f"Total unique schemas: {len(schemas)}")

for i, (schema, files) in enumerate(schemas.items(), 1):
    print(f"\nSchema {i}")
    print(schema[:-3]) # Exclude the metadata columns that we added during data ingestion
    print(f"Columns count: {len(schema) - 3}") # Exclude the metadata columns that we added during data ingestion
    print(f"Files count: {len(files)}")
```
